```python
#Calls the Yahoo finance API to fetch daily stock prices by index
import yfinance as yf
import pandas as pd
import os
```

```python
#Allows download of price history for selected stock
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period = "max")
```

```python
#Pandas dataframe
sp500
```

```python
sp500.index
```

```python
#Plots the trading dates (x-axis) against the trading prices (y-axis) 
sp500.plot.line(y = "Close", use_index = True)
```

```python
#Data trimming, removing extra uneeded columns
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
#Prediction with ML begins here
#Attempting to predict directionality rather than specific prices
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
#Returns a boolean predicting whether tomorrow's price is greater than today's closing price
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype("int")
```

```python
sp500
```

```python
#Removing all data before 1990 to modernize the dataset
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier
#
model = RandomForestClassifier(n_estimators = 100, min_samples_split = 100, random_state = 1)
#n_estimators: Num of decision trees to train, min_samples_split: protects against overfitting, random_state: stops randomness
train = sp500.iloc [:-100]
test = sp500.iloc[-100:]
#Values to be used as predictors. Do not use Tomorrow or Target as the model cannot use this "future" data.
predictors = ["Open", "High", "Low", "Close", "Volume"]
model.fit(train[predictors], train["Target"])
```

```python
#Measuring how accurate the model is (returns a % of the time our prediction was a success)
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
preds = pd.Series(preds, index = test.index)
```

```python
preds
```

```python
precision_score(test["Target"], preds) #Only went up x% of the time
```

```python
combination = pd.concat([test["Target"], preds], axis = 1)
combination.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combination = pd.concat([test["Target"], preds], axis = 1)
    return combination
```

```python
#Constructing a backtesting system
def backtest(data, model, predictors, start = 2500, step = 250):#Tests 10+ years of data
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i + step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
#Addition of more parameters to let the model be more accurate
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_avgs = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_avgs["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
#NaNs are produced when pandas cannot find enough data in the days
sp500 = sp500.dropna()
```

```python
sp500
```

```python
model = RandomForestClassifier(n_estimators = 200, min_samples_split = 50, random_state = 1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    #Has been changed to proba. Returns a % of whether it will return 0 or 1 (stock price is up or down)
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combination = pd.concat([test["Target"], preds], axis = 1)
    return combination
```

```python
#New predictors this time
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
#Our model now says that the stock price goes up between 2 days, 57% of the time
precision_score(predictions["Target"], predictions["Predictions"])
```
